### Tokenisasi

```
# Tokenisasi 
sent_text = nltk.sent_tokenize(lc_par) # tokenisasi menjadi beberapa kalimat
print(sent_text)
# loop per kalimat, tambahkan tag <s> di awal kalimat dan </s> di akhir kalimat
mod_sentences = []
for sentence in sent_text:
    mod_sentence = [] 
    sent_tokens = nltk.word_tokenize(sentence)
    mod_sentence.append('<s>')
    for token in sent_tokens:
      mod_sentence.append(token)
    mod_sentence.append('</s>')
    mod_sentences.append(mod_sentence)
print(mod_sentences)
```

# Unigram

### Membuat tabel frekuensi

```
# Menghitung frekuensi dari setiap kata lalu ditampung pada dictionary freq_tab
freq_tab = {}
total_count = 0
for sentence in mod_sentences:
  for token in sentence:
    if token in freq_tab:
        freq_tab[token] += 1 # kata sudah ada di dictionary, update frekuensinya
    else:
        freq_tab[token] = 1 # kata belum ada di dictionary 
    total_count += 1
        
print(freq_tab)
```

### Membuat tabel probabilitas

```
# Menghitung probabilitas dari setiap kata lalu ditampung pada dictionary prob_tab
prob_tab = {}
for token in freq_tab:
    prob_tab[token] = freq_tab[token]/total_count
    
print(prob_tab)
```

# Bigram

### Membuat tabel frekuensi

```
# Menghitung frekuensi dari setiap kata bigram lalu ditampung pada dictionary freq_bigram_tab
freq_bigram_tab = {}
for sentence in mod_sentences:
  for i in range (1, len(sentence)):
    curr_bigram = (sentence[i-1], sentence[i])
    if curr_bigram in freq_bigram_tab:
      freq_bigram_tab[curr_bigram] += 1 # bigram sudah ada di dictionary, update frekuensinya
    else:
      freq_bigram_tab[curr_bigram] = 1 # bigram belum ada di dictionary
print(freq_bigram_tab)
```

### Membuat tabel probabilitas

```
# Menghitung probabilitas dari setiap kata bigram lalu ditampung pada dictionary bigram_prob_tab
bigram_prob_tab = {}
for sentence in mod_sentences:
  for i in range (1, len(sentence)):
    curr_bigram = (sentence[i-1], sentence[i])

    if curr_bigram not in bigram_prob_tab:
      bigram_prob_tab[curr_bigram] = freq_bigram_tab[curr_bigram]/freq_tab[sentence[i-1]]  # bigram belum ada di dictionary, hitung probability
 
print(bigram_prob_tab)
```

```
# Fungsi laplace smoothing pada unigram
def laplaceunigram(kalimat):
  prob_new_uni = {}
  probnya = 1

  #menghitung keunikan kata dari kalimat uji
  values, counts = np.unique(kalimat, return_counts=True)
  unik = sum(counts)

  for kata in kalimat:
    if kata in freq_tab: #jika kata uji ada di dictionary
      prob_new_uni[kata] = (1 + freq_tab[kata]) / (total_count + unik)  #total_count dari frekuensi unigram
    else: #jika kata uji tidak ada di dictionary
      prob_new_uni[kata] = 1 / (1 + total_count) 
    probnya = probnya * prob_new_uni[kata]
      
  return probnya
```

```
# Fungsi tanpa laplace smoothing pada unigram
def nolaplaceunigram(kalimat):
  probnya = 1
  for kata in kalimat:
    if kata in freq_tab:
      probnya = probnya * prob_tab[kata]
    else: #jika kata tidak ada di tabel frekuensi unigram
      nilai = 0 # 0 karena tidak ditemukan kata di dictionary
      probnya = probnya * nilai
  return probnya
```

# Kalimat Uji Bigram

```
import math
```

```
# Fungsi untuk mengembalikan dictionary frekuensi bigram
def make_freq_bigram(mod_sentences):
  freq_bigram_tab = {}
  for sentence in mod_sentences:
    for i in range (1, len(sentence)):
      curr_bigram = (sentence[i-1], sentence[i])
      if curr_bigram in freq_bigram_tab:
        freq_bigram_tab[curr_bigram] += 1 # bigram sudah ada di dictionary, update frekuensinya
      else:
        freq_bigram_tab[curr_bigram] = 1 # bigram belum ada di dictionary
  return freq_bigram_tab
```

```
# Fungsi laplace smoothing pada bigram
# Mengembalikan nilai probabilitas dari suatu kalimat
def laplacesmoothing(kalimat,unik_count):
  freq_tab_new = {}
  prob = 1
  # Melakukan looping kata pada kalimat
  for sentence in kalimat:
    for i in range(1, len(sentence)):
      curr_bigram = (sentence[i-1], sentence[i])
      if curr_bigram in freq_bigram_tab: #Jika bigram ada di tabel
        freq_tab_new[curr_bigram] = (1 + freq_bigram_tab[curr_bigram]) / (freq_tab[sentence[i-1]] + unik_count)
      else: #Jika bigram tidak ada di tabel
        freq_tab_new[curr_bigram] = 1 / (1 + unik_count)
      prob = prob * freq_tab_new[curr_bigram]
  return prob
```

```
# Fungsi tanpa laplace smoothing pada bigram
# Mengembalikan nilai probabilitas dari suatu kalimat
def nolaplacebigram(kalimat):
  freq_tab_new = {}
  prob = 1
  # Melakukan looping kata pada kalimat
  for sentence in kalimat:
    for i in range(1, len(sentence)):
      curr_bigram = (sentence[i-1], sentence[i])
      if curr_bigram in freq_bigram_tab: #Jika bigram ada di tabel
        prob = prob * bigram_prob_tab[curr_bigram]
      else: #jika kata tidak ada di tabel frekuensi bigram
        nilai = 0 # 0 karena tidak ditemukan kata di dictionary
        prob = prob * nilai
  return prob 
```

```
# Fungsi untuk menghitung perplexity
def perplex(total_prob,N):
  if total_prob != 0:
    # total_prob merupakan probabilitas hasil smothing kalimat
    # N merupakan jumlah jenis kata pada kalimat uji
    px = (total_prob)**(-(1/N))
  else:
    px = 0
  return px
```

### Kalimat Serupa: Bersepeda saat ini mulai kembali tren. Tren bersepeda membuat pabrik mengeluarkan produk baru mereka. Meskipun mahal, sepeda masih laku terjual.

```
# Menguji teks serupa yang tersedia dibawah ini
# Menentukan probabilitasnya dan perplexity per kalimat
test_sentence = 'Bersepeda saat ini mulai kembali tren. Tren bersepeda membuat pabrik mengeluarkan produk baru mereka. Meskipun mahal, sepeda masih laku terjual.'
lc_test_sentence = test_sentence.lower()
kalimat = nltk.sent_tokenize(lc_test_sentence) #tokenisasi menjadi beberapa kalimat

# Pemberian tag <s> dan </s> untuk setiap kalimat dan disimpan di arrat mod_sentences
mod_sentences = []
for sentence in kalimat:
  mod_sentence = []
  sent_tokens = nltk.word_tokenize(sentence)
  mod_sentence.append('<s>')
  for token in sent_tokens:
    mod_sentence.append(token)
  mod_sentence.append('</s>')
  mod_sentences.append(mod_sentence)
print(mod_sentences)
print("------------------------------------------------")

# Menghitung jumlah jenis kata dalam kalimat uji
freq_new = make_freq_bigram(mod_sentences)
unik_count = (len(freq_new))

# Proses perhitungan probabilitas dan perplexity
probsmot = []
probnosmot = []
total_prob_smot = 1
total_prob_nosmot = 1
for i in range(len(mod_sentences)):
  probsmot.append(laplacesmoothing(mod_sentences[i],unik_count))
  probnosmot.append(nolaplacebigram(mod_sentences[i]))

  # Menghitung probabilitas per-kalimat dengan dan tanpa smothint
  total_prob_smot = total_prob_smot * probsmot[i]
  total_prob_nosmot = total_prob_nosmot * probnosmot[i]

  N = len(mod_sentences[i])
  px = perplex(probsmot[i],N) # Parameternya probabilitas kalimat uji
  print("perplex kalimat ke-",i+1,":",px)
  pxno = perplex(probnosmot[i],N)

print("------------------------------------------------")
print("Probabilitas dengan smoothing:", total_prob_smot)
print("Probabilitas tanpa smoothing ", total_prob_nosmot)
```
